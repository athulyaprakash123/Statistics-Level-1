```python
import pandas as pd
import numpy as np
test=pd.DataFrame()
test['a']=pd.Series([34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56 ])
print('mean:',np.mean(test['a']))
test['a-mean']=test['a']-np.mean(test['a'])
test['(a-mean)**2']=test['a-mean']**2
print(test)
print('median:',test['a'].median())
print('sum of squared mean:',sum(test['(a-mean)**2']))
print('count of the data minus one:',len(test['a']-1) )
variance=(sum(test['(a-mean)**2']/len(test['a']-1)))
print('variance:',variance)
st_dev=np.sqrt(variance)
print( 'standard deviation:',st_dev)
```
